 Firstly, clean the years names and make them numerical

```python
#Clean Years

raw_data=raw_data.rename({ "2001-02": "2001", "2002-03": "2002", "2003-04": "2003", "2004-05": "2004", "2005-06": "2005", "201516": "2016", "201617": "2017", "201718": "2018", "201819":"2019", "201920":"2020"}, axis=1)



all_years = raw_data.iloc[:, 6:54]
all_years = all_years.replace(',','', regex=True)
all_years= all_years.apply(pd.to_numeric, errors='coerce')
all_years = all_years.round().astype('Int64')


all_years
```

```python
#Only interested in last 10 years
N=10
last_10_years=all_years.iloc[: , -N:]

last_10_years
```

```python
#Check for NA's. Makes sense to fill them with 0.
last_10_years.isna().sum()
last_10_years=last_10_years.fillna(0)

#print(last_10_years.isna().sum())
```

```python
#Add Food Code

food_code_column = raw_data["Code"]

last_10_years = pd.concat([last_10_years,food_code_column], axis = 1)

food_code_col_name="Code"
first_col_all_years = last_10_years.pop("Code")
last_10_years.insert(0, food_code_col_name, first_col_all_years )
```

```python
#Add units

#Add Units column

units_column = raw_data["Units"]
last_10_years = pd.concat([last_10_years,units_column], axis = 1)

units_column_name="Units"
sec_col_all_years = last_10_years.pop("Units")
last_10_years.insert(1, units_column_name, sec_col_all_years )
last_10_years
```

```python
raw_food_groups=raw_data.iloc[:, 0:5]
raw_food_groups
```

```python
#Keep only high level food categories

key_food_groups=raw_food_groups.dropna(subset=['Food Category', 'Food Group'], how='all').drop(columns=["Major Food Code", "Minor Food Code"])
key_food_groups
```

```python
# Make Food Catefories one column. np.where(condition, if true, if false)
key_food_groups['Food Category'] = np.where(key_food_groups['Food Category'].isna(),key_food_groups['Food Group'], key_food_groups['Food Category'])
key_food_groups=key_food_groups.drop(key_food_groups.columns[2], axis=1)
key_food_groups
```

```python
#Remove smaller food groups from this list
remove_group_codes=["4006","9017","NaN","cat811","162171","172183","cat702","cat730", "cat810", "251263"]
remove_group_codes
key_food_groups=key_food_groups[~key_food_groups['Code'].isin(remove_group_codes)].fillna('missing code')

key_food_groups
#If time! Give the codes all category or minor category labels. Definitely replace the NaN
#Add clearer Food Cat labels if time
```

```python
#Merge food categories and years
household_food_consumption_clean=key_food_groups.merge(last_10_years, left_on='Code', right_on='Code', how='inner')
household_food_consumption_clean.sort_values(by=['2020', '2019'], ascending=False).reset_index(inplace=True)
household_food_consumption_clean
```

```python
(household_food_consumption_clean.iloc[1,1] in ['Cheese'])
```

```python
for row in range(household_food_consumption_clean.shape[0]):
    if (household_food_consumption_clean.iloc[row,1] in ['Milk and milk products excluding cheese'])==True:
         household_food_consumption_clean['Main Group'] = "Milk products exc. cheese"
household_food_consumption_clean
```

```python
def clean_categories(x):
    if (x in ['Milk and milk products excluding cheese']):
        return "Milk products exc. cheese"
    elif (x in ['Cheese']):
        return "Cheese"
    elif (x in ['Carcase meat','Non-carcase meat and meat products']):
        return "Meat"
    elif (x in ['Fish']):
        return "Fish"
    elif (x in ['Eggs']):
        return "Eggs"
    elif (x in ['Fats']):
        return "Fats"
    elif (x in ['Sugar and preserves']):
        return "Sugar and preserves"
    elif (x in ['Fresh and processed potatoes','Processed potatoes']):
        return "Potatoes"
    elif (x in ['Fresh and processed vegetables, excluding potatoes']):
        return "Vegetables exc. Potatoes"
    elif (x in ['Fresh fruit','Processed fruit and fruit products']):
        return "Fruit"
    elif (x in ['White bread','Brown and wholemeal bread']):
        return "Bread"
    elif (x in ['Flour']):
        return "Flour"
    elif (x in ['Cakes, buns and pastries']):
        return "Cakes, buns and pastries"
    elif (x in ['Biscuits and crispbreads']):
        return "Biscuits"
    elif (x in ['Other cereals and cereal products']):
        return "Other cereals and cereal products"
    elif (x in ['Beverages']):
        return "Tea and Coffee"
    elif (x in ['Other food and drink']):
        return "Other food and drink"
    elif (x in ['Soft drinks']):
        return "Soft drinks"
    elif (x in ['Confectionery']):
        return "Confectionery"
    elif (x in ['Alcoholic drinks']):
        return "Alcoholic drinks"
    else: return "Minor food group"
    
```

```python
household_food_consumption_clean['Main Group'] = household_food_consumption_clean['Food Category'].apply(clean_categories)

main_group_col = household_food_consumption_clean.pop("Main Group")
household_food_consumption_clean.insert(1, "Main Group", main_group_col)
#.drop(key_food_groups.columns[2], axis=1)
household_food_consumption_clean=household_food_consumption_clean.drop(['Code'], axis=1)

household_food_consumption_clean
```

```python
melted = household_food_consumption_clean.melt(id_vars=['Main Group','Food Category'],
                                      value_vars = ['2011','2012','2013','2014','2015','2016',
                                                    '2017','2018','2019','2020']).rename(columns={'variable':'year'})
melted.to_csv("melted.csv",index= False)
```
